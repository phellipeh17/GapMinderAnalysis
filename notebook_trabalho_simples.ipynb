```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Vou criar o Bloco principal que vai carregar os dados


```python
if __name__ == "__main__":
  
    df = carregar_dados('gapminder_full.csv')
```

No próximo bloco realizei a Leitura e Tratamento:
- Carreguei o CSV e retorna o DataFrame.
- Filtrei o DataFrame para um ano específico.

```python
def carregar_dados(caminho_arquivo):
    return pd.read_csv(caminho_arquivo)

def filtrar_por_ano(df, ano):
    return df[df['year'] == ano].copy()
```

No próximo modulo criei as funções de gráficos (Cada gráfico fiz uma função) para ser chamada quando necessária

```python

def grafico_evolucao_global(df):
    #Gráfico 1: Média global ao longo do tempo.
    media_ano = df.groupby('year')['lifeExp'].mean().reset_index()
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=media_ano, x='year', y='lifeExp', marker='o')
    plt.title('Evolução da Expectativa de Vida Global')
    plt.ylabel('Anos')
    plt.show()


def grafico_evolucao_continente(df):
    """Gráfico 2: Comparação entre continentes."""
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=df, x='year', y='lifeExp', hue='continent', estimator='mean')
    plt.title('Evolução por Continente')
    plt.show()

def grafico_distribuicao_box(df, ano=2007):
    """Gráfico 3: Boxplot para ver a dispersão/desigualdade."""
    df_ano = filtrar_por_ano(df, ano)
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df_ano, x='continent', y='lifeExp')
    plt.title(f'Distribuição da Expectativa de Vida ({ano})')
    plt.show()

def grafico_scatter_pib(df, ano=2007):
    """Gráfico 4: Correlação com Riqueza (PIB)."""
    df_ano = filtrar_por_ano(df, ano)
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=df_ano, x='gdpPercap', y='lifeExp', hue='continent', size='pop', alpha=0.7)
    plt.xscale('log') # Escala log para melhor visualização
    plt.title(f'Expectativa de Vida vs PIB ({ano})')
    plt.show()

def grafico_ranking_top(df, ano=2007):
    """Gráfico 5: Top 10 países."""
    df_ano = filtrar_por_ano(df, ano)
    top10 = df_ano.nlargest(10, 'lifeExp')
    plt.figure(figsize=(10, 6))
    sns.barplot(data=top10, x='lifeExp', y='country')
    plt.title(f'Top 10 Países em Expectativa de Vida ({ano})')
    plt.xlim(70, 90) # Foca na parte superior da escala
    plt.show()
```

Gerar visualizações chamando as funções e explicação dos insights

```python
# Chamar a função principal para gerar o gráfico
grafico_evolucao_global(df)
```

Ao olhar para a média global, fica bem claro que o mundo progrediu muito nas últimas décadas. A linha é ascendente e constante desde 1952, o que sugere que avanços na medicina, saneamento e acesso a alimentos impactaram positivamente a longevidade humana de forma geral. Não houve quedas bruscas na média mundial, indicando uma melhoria contínua.
